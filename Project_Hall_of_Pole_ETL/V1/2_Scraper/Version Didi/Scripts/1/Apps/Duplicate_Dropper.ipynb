Function

```python
import pandas as pd

def drop_duplicates_and_merge(df1, df2, on_col):
    url_mapping = df2.loc[:, [on_col, df2.columns[-1]]]
    merged_df = pd.merge(df1, url_mapping, on=on_col, how='outer')
    merged_df = merged_df.drop_duplicates(subset=on_col)
    merged_df = merged_df.reset_index(drop=True)
    return merged_df
```

```python
df1 = pd.read_csv('all_pole_studios_germany.csv')
df2 = pd.read_csv('all_pole_studios_germany_cl.csv')
merged_df = drop_duplicates_and_merge(df1, df2, 'Business Name')
merged_df
```
