```python
import pandas as pd
from datetime import datetime
import requests
```

```python
valid_urls = pd.read_csv("valid_eversports_urls_didi.csv")
valid_urls
```

---------

URL Sorter SW, S, SK

```python
def sort_urls(valid_urls):
    sorted_urls = {}
    sorted_urls['eversports_de_sw'] = valid_urls[valid_urls['Valid_URL'].str.contains('https://www.eversports.de/sw/')]
    sorted_urls['eversports_de_s'] = valid_urls[valid_urls['Valid_URL'].str.contains('https://www.eversports.de/s/')]
    sorted_urls['eversports_de_sk'] = valid_urls[valid_urls['Valid_URL'].str.contains('https://www.eversports.de/sk/')]
    # add more categories as needed
    return sorted_urls
```

```python
sorted_urls = sort_urls(valid_urls)
sorted_urls
```

```python
# Create empty dataframes for each category of URL
sw_urls = pd.DataFrame(columns=["Valid_URL"])
s_urls = pd.DataFrame(columns=["Valid_URL"])
sk_urls = pd.DataFrame(columns=["Valid_URL"])

# Loop through each URL in the valid_urls dataframe
for i in range(len(valid_urls)):
    url = valid_urls.loc[i, "Valid_URL"]
    
    # Check the URL category and append to the appropriate dataframe
    if "/sw/" in url:
        sw_urls = sw_urls.append({"Valid_URL": url}, ignore_index=True)
    elif "/s/" in url:
        s_urls = s_urls.append({"Valid_URL": url}, ignore_index=True)
    elif "/sk/" in url:
        sk_urls = sk_urls.append({"Valid_URL": url}, ignore_index=True)
        
# Print out the resulting dataframes
print("SW URLs:")
print(sw_urls)
print("\nS URLs:")
print(s_urls)
print("\nSK URLs:")
print(sk_urls)
```

```python
# Concatenate the DataFrames horizontally, using the column names as keys
all_valid_urls = pd.concat([sw_urls, s_urls, sk_urls], axis=1, keys=['Valid_URL', 'Valid_URL', 'Valid_URL'])

# Concatenate the dataframes while keeping their original column names
all_valid_urls = pd.concat([sw_urls.rename(columns={"Valid_URL": "sw"}),
                            s_urls.rename(columns={"Valid_URL": "s"}),
                            sk_urls.rename(columns={"Valid_URL": "sk"})], axis=1)
all_valid_urls
```
